```python
#old dataset (misnamed 16 but it's actually 12)
df16 = pd.read_excel('oesm12st/state_M2012_dl.xls' , na_values = ['#', '**', '*'], converters={u'TOT_EMP':float, u'EMP_PRSE':float, u'JOBS_1000':float, u'LOC_Q':float, u'H_MEAN':float, u'A_MEAN':float,
       u'MEAN_PRSE':float, u'H_PCT10':float, u'H_PCT25':float, u'H_MEDIAN':float, u'H_PCT75':float,
       u'H_PCT90':float, u'A_PCT10':float, u'A_PCT25':float, u'A_MEDIAN':float, u'A_PCT75':float, u'A_PCT90':float})
```

```python
#new dataset
df17 = pd.read_excel('oesm18st/state_M2018_dl.xlsx', na_values = ['#', '**', '*'], converters={u'TOT_EMP':float, u'EMP_PRSE':float, u'JOBS_1000':float, u'LOC_Q':float, u'H_MEAN':float, u'A_MEAN':float,
       u'MEAN_PRSE':float, u'H_PCT10':float, u'H_PCT25':float, u'H_MEDIAN':float, u'H_PCT75':float,
       u'H_PCT90':float, u'A_PCT10':float, u'A_PCT25':float, u'A_MEDIAN':float, u'A_PCT75':float, u'A_PCT90':float})
```

```python
#filtered to NJ
nj16 = df16[df16.ST == 'NJ']
```

```python
#filtered to NJ
nj17 = df17[df17.ST == 'NJ']
```

```python
#merging old and new
nj1617 = nj17.merge(nj16, how='outer', on='OCC_CODE', suffixes=['_17', '_12'])
```

```python
nj1617.columns
```

```python
#calculating the inflation-adjusted salary change
nj1617['12_amed_INF_ADJ'] = nj1617.A_MEDIAN_12 + nj1617.A_MEDIAN_12 * .0687
```

```python
#getting the percent changes
nj1617['JOB_CHANGE'] = ((nj1617['TOT_EMP_17'] - nj1617.TOT_EMP_12) / nj1617.TOT_EMP_12) *100
nj1617['WAGE_CHANGE'] = ((nj1617['A_MEDIAN_17'] - nj1617['12_amed_INF_ADJ'] ) / nj1617['12_amed_INF_ADJ'] ) *100
```
